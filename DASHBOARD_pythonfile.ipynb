```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Department': ['Sales', 'Research & Development', 'HR', 'IT'],
    'Number of Employees': [300, 500, 50, 100]
})

# Define a custom color palette with distinct colors
palette = sns.color_palette("husl", len(data))

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Number of Employees', data=data, palette=palette)

# Add labels and title
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees by Department with Distinct Colors')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data['Age'], data['DailyRate'], data['DistanceFromHome'], c=data['Attrition'].astype('category').cat.codes, cmap='viridis')
ax.set_xlabel('Age')
ax.set_ylabel('DailyRate')
ax.set_zlabel('DistanceFromHome')
ax.set_title('3D Scatter Plot')
fig.colorbar(scatter, ax=ax, label='Attrition')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Create a sample dataset with 4 dimensions
np.random.seed(0)
data = pd.DataFrame({
    'X': np.random.uniform(0, 100, 100),
    'Y': np.random.uniform(0, 100, 100),
    'Z': np.random.uniform(0, 100, 100),
    'Color': np.random.uniform(0, 100, 100),  # 4th Dimension
    'Size': np.random.uniform(10, 100, 100)   # 5th Dimension
})

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with X, Y, Z, Color (used for color map), and Size (used for marker size)
sc = ax.scatter(data['X'], data['Y'], data['Z'], c=data['Color'], s=data['Size'], cmap='viridis', alpha=0.7, edgecolors='w')

# Add color bar
cbar = plt.colorbar(sc, pad=0.1)
cbar.set_label('Color Dimension')

# Labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.title('4D Scatter Plot using Matplotlib')

plt.show()
```

```python
import plotly.express as px

fig = px.scatter_3d(data, x='Age', y='DailyRate', z='DistanceFromHome',
                    color='Education', size='TotalWorkingYears', hover_name='Attrition',
                    title='5D Scatter Plot (Color: Education, Size: TotalWorkingYears)')
fig.update_traces(marker=dict(size=5, opacity=0.7))
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation (replace with your actual data loading code)
data = pd.DataFrame({
    'BusinessTravel': ['Travel_Rarely', 'Travel_Rarely', 'Travel_Frequently', 'Non-Travel', 'Non-Travel']
})

# Calculate proportions
business_travel_counts = data['BusinessTravel'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(business_travel_counts, labels=business_travel_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'), startangle=140)

# Add title
plt.title('Proportion of Business Travel Categories')

plt.show()
```
